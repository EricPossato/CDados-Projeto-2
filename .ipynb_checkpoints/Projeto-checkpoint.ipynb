```python
import numpy as np
import pandas as pd 
import plotly.express as px
```

```python
raw_dataset = pd.read_csv('dataset.csv')
```

```python
print(list(raw_dataset.columns))
colunas_relevantes = raw_dataset[['registrou_bo', 'sexo', 'tipo_assalto_id', 'Bicicleta', 'Bolsa ou Mochila', 'Carteira', 'Cartão de Crédito', 'Celular', 'Computador', 'DVD', 'Dinheiro', 'Documentos', 'Equipamento de Som', 'Estepe', 'MP4 ou Ipod', 'Móveis', 'Notebook', 'Outros', 'Relógio', 'Som', 'Tablet', 'Tv']]
colunas_relevantes.head()
```

```python
#Limpar se registrou o BO
dataset_limpo = colunas_relevantes.loc[(colunas_relevantes['registrou_bo'] == True) | (colunas_relevantes['registrou_bo'] == False)]
#Limpar sexo
dataset_limpo = dataset_limpo.loc[dataset_limpo['sexo'] >= 0]
#Limpar ID tipo de assalto
dataset_limpo = dataset_limpo.loc[dataset_limpo['tipo_assalto_id'] >= 0]


dataset_limpo = dataset_limpo.fillna(False)
dataset_limpo
nulo = dataset_limpo.isnull().sum()
```

```python
print(nulo) #Confirmando se está limpo
```

```python
#separando variáveis para tabela crusada
registrou_bo = dataset_limpo['registrou_bo'] 
sexo = dataset_limpo['sexo'] 

```

```python
pd.crosstab(registrou_bo, sexo,normalize = 'columns')
```

```python
lista_var = list(dataset_limpo.columns)
lista_var.remove('registrou_bo')
lista_var.remove('sexo')
lista_var.remove('tipo_assalto_id')

lista_var 
```

```python
for var in lista_var:
    display(pd.crosstab(registrou_bo, dataset_limpo[var],normalize = 'columns'))
```
